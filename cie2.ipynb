```python
#Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Perform EDA
print(titanic.head())
print(titanic.info())
print(titanic.describe())

# Find missing values
print("Missing values:\n", titanic.isnull().sum())

# Missing values for Survived and Age
print("Missing in Survived:", titanic['survived'].isnull().sum())
print("Missing in Age:", titanic['age'].isnull().sum())

# Bar graph for passenger class
sns.countplot(data=titanic, x='pclass')
plt.title("Passenger Count by Class")
plt.show()
```
